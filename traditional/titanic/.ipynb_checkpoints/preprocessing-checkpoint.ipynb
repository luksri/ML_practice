```python
import pandas as pd
import os 
from sklearn import preprocessing
from sklearn import impute
import seaborn as sns
```

```python
dir = "C:/Users/pc/Downloads/ai-level1/titanic"
titanic_train = pd.read_csv(os.path.join(dir, "train.csv"))
titanic_train.shape
titanic_train.info()
```

```python
features = ['Age', 'Embarked', 'Fare']
imp = impute.SimpleImputer(strategy="most_frequent")
imp.fit(titanic_train[features])
print(imp.statistics_)
titanic_train[features] = imp.transform(titanic_train[features])
```

```python
titanic_train.info()
```

```python
#it encodes a single categorical feature
le = preprocessing.LabelEncoder()
le.fit(titanic_train['Survived'])
print(le.classes_)
titanic_train['Survived'] = le.transform(titanic_train['Survived'])
titanic_train.head()
```

```python
features = ['Pclass']
oe = preprocessing.OrdinalEncoder()
oe.fit(titanic_train[features])
print(oe.categories_)
titanic_train[['Pclass1']] = oe.transform(titanic_train[features])
titanic_train.head()
```

```python
features = ['Pclass']
order = [[1, 2, 3]]
oe = preprocessing.OrdinalEncoder(categories=order)
oe.fit(titanic_train[features])
print(oe.categories_)
titanic_train[['Pclass2']] = oe.transform(titanic_train[features])
titanic_train.head()
```

```python
features = ['Sex', 'Embarked']
ohe = preprocessing.OneHotEncoder(sparse_output=False)
ohe.fit(titanic_train[features])
print(ohe.categories_)
tmp = ohe.transform(titanic_train[features])
df = pd.DataFrame(tmp, columns = ohe.get_feature_names_out(features))
df.info()
```

```python
titanic_train['Fare'] = titanic_train['Fare'].astype('float')
sns.FacetGrid(titanic_train, hue="Survived").map(sns.kdeplot, "Fare").add_legend()
```

```python
titanic_train['Age'] = titanic_train['Age'].astype('float')
sns.FacetGrid(titanic_train, hue="Survived").map(sns.kdeplot, "Age").add_legend()
```

```python
features = ['Fare', 'Age']
kd = preprocessing.KBinsDiscretizer(n_bins=[13, 10], encode='ordinal', strategy='quantile')
kd.fit(titanic_train[features])
print(kd.n_bins_)
print(kd.bin_edges_)
titanic_train[['Fare1', 'Age1']] = kd.transform(titanic_train[features])
titanic_train.head()
```

```python
g = sns.catplot(x="Fare1", hue="Survived", data=titanic_train, kind="count")
g.set_xticklabels(rotation=90)
```

```python
g = sns.catplot(x="Age1", hue="Survived", data=titanic_train, kind="count")
g.set_xticklabels(rotation=90)
```

```python
features = ['Fare', 'Age']
scaler = preprocessing.StandardScaler()
scaler.fit(titanic_train[features])
print(scaler.mean_)
print(scaler.scale_)
titanic_train[['Fare_ss', 'Age_ss']] = scaler.transform(titanic_train[features])
titanic_train.head()
```

```python
scaler = preprocessing.RobustScaler()
scaler.fit(titanic_train[features])
print(scaler.center_)
print(scaler.scale_)
titanic_train[['Fare_rs', 'Age_rs']] = scaler.transform(titanic_train[features])
titanic_train.head()
```

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(titanic_train[features])
print(scaler.min_)
print(scaler.scale_)
titanic_train[['Fare_ms', 'Age_ms']] = scaler.transform(titanic_train[features])
titanic_train.head()
```
