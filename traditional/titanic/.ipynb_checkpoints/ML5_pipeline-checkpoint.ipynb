```python
import os
import pandas as pd
import numpy as np
import sklearn
from sklearn import tree
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import impute
from sklearn import pipeline
import math
sklearn.set_config(display="diagram")
```

```python
dir = "."
tr_data = pd.read_csv(os.path.join(dir,'train.csv'))
tr_data.head()
```

```python
#### create the pipeline instance with two stages defined. one is to impute- replace missing values
#.    and other is model creation.
steps = [('imp', impute.SimpleImputer()), ('dt', tree.DecisionTreeClassifier())]
pipe = pipeline.Pipeline(steps)
```

```python
pipe
```

```python
#### now attach parameter tuning to the pipeline parameter grid with stage info.
pipe_grid = {'dt__max_depth':list(range(1,11)),'dt__min_samples_split':[2,5,10], 'imp__strategy':['mean', 'median']}
cv = model_selection.KFold(n_splits=10)
clf = model_selection.GridSearchCV(pipe, pipe_grid, cv=cv, scoring='accuracy',return_train_score=True)

## select the features and train the model
features = ['SibSp', 'Parch', 'Age']
X_train = tr_data[features]
y_train = tr_data['Survived']
clf.fit(X_train, y_train)

### explore the results from the trained model

print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
print(type(clf.best_estimator_), clf.best_estimator_.score(X_train, y_train))
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python

test_data = pd.read_csv(os.path.join(dir,'test.csv'))

### now we dont need to encode the test data. we can simply call the model.
## when we use object 'clf' --> it will take everything through pipeline. no need to explicitly write transform.

### question: on train data -> fit and transform for features, on test - only transform. how pipeline is ensuring?


# test_data['Sex_encoded']=sex_le.transform(test_data['Sex'])
# test_data['Embarked_encoded']=embarked_le.transform(test_data['Embarked'])
# test_data['Age_encoded']=age_imp.transform(test_data[['Age']])

X_test = test_data[features]


### predict the output for the test data
test_data['Survived']=clf.predict(X_test)
test_data.head()
```
