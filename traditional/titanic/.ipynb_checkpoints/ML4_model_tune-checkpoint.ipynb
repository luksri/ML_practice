```python
import os
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import impute
import math
```

```python
dir = "."
tr_data = pd.read_csv(os.path.join(dir,'train.csv'))
tr_data.head()
```

```python
### clean the data now. to do this, we need to convert the data to canonical form so that we can decision trees(requirement by the algo)
sex_le = preprocessing.LabelEncoder()
tr_data['Sex_encoded']=sex_le.fit_transform(tr_data['Sex'])
tr_data.head()
```

```python
embarked_le = preprocessing.LabelEncoder()
tr_data['Embarked_encoded']=embarked_le.fit_transform(tr_data['Embarked'])
tr_data.head()
```

```python
age_imp = impute.SimpleImputer()
tr_data['Age_encoded']=age_imp.fit_transform(tr_data[['Age']])
tr_data.head()
```

```python
### feature engineering//feature selection
features = ['Parch','SibSp', 'Sex_encoded','Embarked_encoded', 'Age_encoded']
X_train = tr_data[features]
y_train = tr_data['Survived']
```

```python
#### data is preprocessed till now. now time for model creation and parameter tuning for accuracy
dt = tree.DecisionTreeClassifier()
dt_param_grid = {'max_depth':list(range(1,11)),'min_samples_split':[2,5,10]}
cv = model_selection.KFold(n_splits=10)

clf = model_selection.GridSearchCV(dt, dt_param_grid, cv=cv, scoring='accuracy',return_train_score=True)

#learn the pattern in the data by invoking fit method
clf.fit(X_train,y_train)

print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python
#### data is preprocessed till now. now time for model creation and parameter tuning for accuracy
dt = tree.DecisionTreeClassifier()
dt_param_grid = {'max_depth':list(range(1,11)),'min_samples_split':[2,5,10], 'max_features':list(range(1, len(features)+1)), 'criterion':['gini', 'entropy']}
cv = model_selection.KFold(n_splits=10)

clf = model_selection.GridSearchCV(dt, dt_param_grid, cv=cv, scoring='accuracy',return_train_score=True)

#learn the pattern in the data by invoking fit method
clf.fit(X_train,y_train)

print(clf.best_params_)
print(clf.best_score_)  ## this is the cross val score of all folds
print(clf.best_index_)
print(clf.best_estimator_)
print(clf.best_estimator_.score(X_train,y_train))   ### check the scores to understand overfit or underfit on the actual train that we used on model creation
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python

test_data = pd.read_csv(os.path.join(dir,'test.csv'))
test_data['Sex_encoded']=sex_le.transform(test_data['Sex'])
test_data['Embarked_encoded']=embarked_le.transform(test_data['Embarked'])
test_data['Age_encoded']=age_imp.transform(test_data[['Age']])


X_test = test_data[features]


### predict the output for the test data
test_data['Survived']=clf.best_estimator_.predict(X_test)
test_data.head()
```

```python
test_data.to_csv(os.path.join(dir, "submission1.csv"), columns=["PassengerId", "Survived"], index=False)
```

```python
import joblib

#persist model using joblib
objects = {
    'features':features,
    'le_sex': sex_le,
    'le_emb': embarked_le,
    'imp_age': age_imp,
    'clf':clf
}
joblib.dump(objects, os.path.join(dir, "titanic_v1.pkl"))
```
