```python
import pandas as pd
import os
import seaborn as sns
```

```python
dir = "C:/Users/pc/Downloads/ai-level1/titanic"
titanic_train = pd.read_csv(os.path.join(dir, "train.csv"))
titanic_train.shape
titanic_train.info()
```

```python
#explore bivariate relationships: categorical vs categorical 
sns.catplot(x="Sex", hue="Survived", data=titanic_train, kind="count")
```

```python
sns.catplot(x="Pclass", hue="Survived", data=titanic_train, kind="count")
```

```python
sns.catplot(x="Embarked", hue="Survived", data=titanic_train, kind="count")
```

```python
#explore bivariate relationships: continuous  vs categorical
sns.FacetGrid(titanic_train, hue="Survived").map(sns.kdeplot, "Age").add_legend()
```

```python
sns.FacetGrid(titanic_train, hue="Survived").map(sns.kdeplot, "Fare").add_legend()
```

```python
sns.FacetGrid(titanic_train, hue="Survived").map(sns.kdeplot, "Fare").add_legend()
```

```python
sns.FacetGrid(titanic_train, col="Survived").map(sns.boxplot, "Fare").add_legend()
```

```python
#explore bivariate relationships: continuous vs continuous 
sns.jointplot(x="Age", y="Fare", data=titanic_train)
```
