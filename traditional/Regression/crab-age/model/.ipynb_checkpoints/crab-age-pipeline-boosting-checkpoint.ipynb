```python
import pandas as pd
import os
import sklearn
import math
import joblib
from sklearn import tree
from sklearn import ensemble 
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import impute
from sklearn import pipeline
from sklearn import compose
from sklearn import feature_selection
from sklearn import neighbors
from sklearn import linear_model
from sklearn import svm
from sklearn import ensemble
from sklearn import metrics
from sklearn.compose import TransformedTargetRegressor
from feature_engine.selection import DropDuplicateFeatures, DropCorrelatedFeatures
import numpy as np
import xgboost as xgb
import lightgbm as lgb
# import catboost as cb
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from shutil import rmtree
#pip install xgboost
#pip install catboost
#pip install lightgbm
```

```python
warnings.filterwarnings('ignore')
sklearn.set_config(display="diagram")
dir = "../data"
cachedir = os.path.join(dir, "pipeline_cache")
```

```python
train_data = pd.read_csv(os.path.join(dir, "train.csv"))
train_data.shape
train_data.info()
```

```python
def cont_selector(df):
    return df.select_dtypes(include=['number']).columns

def cat_selector(df):
    return df.select_dtypes(exclude=['number']).columns

def cast_to_cat(df, features):
    for feature in features:
        df[feature] = df[feature].astype('category')
```

```python
print(cont_selector(train_data))
print(cat_selector(train_data))
```

```python
def get_features_to_drop_on_missingdata(df, threshold) :
    tmp = df.isnull().sum()
    return list(tmp[tmp/float(df.shape[0]) > threshold].index)

def drop_features(df, features):
    return df.drop(features, axis=1, inplace=True)
```

```python
missing_data_features_to_drop = get_features_to_drop_on_missingdata(train_data, 0.25)
print(missing_data_features_to_drop)
drop_features(train_data, missing_data_features_to_drop)
train_data.info()
```

```python
target = train_data['Age']
features_to_drop = ['id', 'Age']
drop_features(train_data, features_to_drop)
train_data.info()
```

```python
#define train, target data and metric
X_train = train_data
y_train = target
scoring = metrics.make_scorer(metrics.mean_absolute_error)
target_transformer = preprocessing.PowerTransformer(method='box-cox')
```

```python
#preprocessing pipeline for tree estimators
steps = [('imp', impute.SimpleImputer(strategy="most_frequent")), 
         ('ord_enc', preprocessing.OrdinalEncoder(
                        handle_unknown="use_encoded_value",
                        unknown_value=-1,
                        encoded_missing_value=-2))
        ]
cat_tree_preprocessor = pipeline.Pipeline(steps)

steps = [('imp', impute.SimpleImputer())]
cont_tree_preprocessor = pipeline.Pipeline(steps)

tree_preprocessor = compose.ColumnTransformer(
    [ ("categorical", cat_tree_preprocessor, cat_selector(train_data)), ("continuous", cont_tree_preprocessor, cont_selector(train_data)) ]
)
```

```python
#hgb pipeline-I
hgb_pipeline1 = pipeline.Pipeline([  
                    ('tree_preprocessor', tree_preprocessor),
                    ('zv_filter', feature_selection.VarianceThreshold()),
                    ('hgb', ensemble.HistGradientBoostingRegressor())
                ], memory=cachedir)
hgb_pipeline1
```

```python
#grid search of hgb pipeline-I
hgb_pipeline_params1 = {
              'hgb__learning_rate': [1, 0.25, 0.1, 0.05, 0.01],
              'hgb__max_depth': [3, 4, 5, 7],
              'hgb__min_samples_leaf': [1, 3, 10],
              'hgb__max_features': [0.5, 0.3, 0.1]
              }
cv = model_selection.KFold(10)
hgb_pipeline_grid1 = model_selection.GridSearchCV(hgb_pipeline1, hgb_pipeline_params1, cv=cv, scoring=scoring, n_jobs=-1)
hgb_pipeline_grid1.fit(X_train, y_train)
print(hgb_pipeline_grid1.best_params_)
print(hgb_pipeline_grid1.best_score_)
print(hgb_pipeline_grid1.best_estimator_)
```

```python
#hgb pipeline-II
hgb_with_trans_target = TransformedTargetRegressor(
    regressor = ensemble.HistGradientBoostingRegressor(),
    transformer= target_transformer
)
hgb_pipeline2 = pipeline.Pipeline([  
                    ('tree_preprocessor', tree_preprocessor),
                    ('zv_filter', feature_selection.VarianceThreshold()),
                    ('hgb', hgb_with_trans_target)
                ], memory=cachedir)
hgb_pipeline2
```

```python
#grid search of hgb pipeline-II
hgb_pipeline_params2 = {
              'hgb__regressor__learning_rate': [1, 0.25, 0.1, 0.05, 0.01],
              'hgb__regressor__max_depth': [3, 4, 5, 7],
              'hgb__regressor__min_samples_leaf': [1, 3, 10],
              'hgb__regressor__max_features': [0.5, 0.3, 0.1]
              }
cv = model_selection.KFold(10)
hgb_pipeline_grid2 = model_selection.GridSearchCV(hgb_pipeline2, hgb_pipeline_params2, cv=cv, scoring=scoring, n_jobs=-1)
hgb_pipeline_grid2.fit(X_train, y_train)
print(hgb_pipeline_grid2.best_params_)
print(hgb_pipeline_grid2.best_score_)
print(hgb_pipeline_grid2.best_estimator_)
```

```python
#xgb pipeline-I
xgb_pipeline1 = pipeline.Pipeline([  
                    ('tree_preprocessor', tree_preprocessor),
                    ('zv_filter', feature_selection.VarianceThreshold()),
                    ('xgb', xgb.XGBRegressor())
                ], memory=cachedir)
xgb_pipeline1
```

```python
#grid search of xgb pipeline-I
min_child_weight = np.linspace(1, 10, 5, endpoint=True) 
gamma = np.linspace(0.5, 5, 5, endpoint=True)
subsample = np.linspace(0.5, 1, 5, endpoint=True)
colsample_bytree = np.linspace(0.5, 1, 5, endpoint=True)
xgb_pipeline_params1 = {
        'xgb__min_child_weight': min_child_weight,
        'xgb__gamma': gamma,
        'xgb__subsample': subsample,
        'xgb__colsample_bytree': colsample_bytree,
        'xgb__max_depth': [3,5,7]
        } 

cv = model_selection.KFold(10)
xgb_pipeline_grid1 = model_selection.GridSearchCV(xgb_pipeline1, xgb_pipeline_params1, cv=cv, scoring=scoring, n_jobs=4)
xgb_pipeline_grid1.fit(X_train, y_train)
print(xgb_pipeline_grid1.best_params_)
print(xgb_pipeline_grid1.best_score_)
print(xgb_pipeline_grid1.best_estimator_)
```

```python
#xgb pipeline-II
xgb_with_trans_target = TransformedTargetRegressor(
    regressor = xgb.XGBRegressor(),
    transformer= target_transformer
)
xgb_pipeline2 = pipeline.Pipeline([  
                    ('tree_preprocessor', tree_preprocessor),
                    ('zv_filter', feature_selection.VarianceThreshold()),
                    ('xgb', xgb_with_trans_target)
                ], memory=cachedir)
xgb_pipeline2
```

```python
#grid search of xgb pipeline-II
min_child_weight = np.linspace(1, 10, 5, endpoint=True) 
gamma = np.linspace(0.5, 5, 5, endpoint=True)
subsample = np.linspace(0.5, 1, 5, endpoint=True)
colsample_bytree = np.linspace(0.5, 1, 5, endpoint=True)
xgb_pipeline_params2 = {
        'xgb__regressor__min_child_weight': min_child_weight,
        'xgb__regressor__gamma': gamma,
        'xgb__regressor__subsample': subsample,
        'xgb__regressor__colsample_bytree': colsample_bytree,
        'xgb__regressor__max_depth': [3,5,7]
        } 

cv = model_selection.KFold(10)
xgb_pipeline_grid2 = model_selection.GridSearchCV(xgb_pipeline2, xgb_pipeline_params2, cv=cv, scoring=scoring, n_jobs=4)
xgb_pipeline_grid1.fit(X_train, y_train)
print(xgb_pipeline_grid2.best_params_)
print(xgb_pipeline_grid2.best_score_)
print(xgb_pipeline_grid2.best_estimator_)
```

```python
feature_names = xgb_pipeline_grid2.best_estimator_[0].get_feature_names_out()
print(len(feature_names), feature_names)
feature_names = xgb_pipeline_grid2.best_estimator_[1].get_feature_names_out(input_features = feature_names)
print(len(feature_names), feature_names)
```

```python
def plot_feature_importances(model, feature_names):
    features = pd.DataFrame({'feature': feature_names, 'importance':model.feature_importances_})
    features.sort_values(by=['importance'], ascending=True, inplace=True)
    features.set_index('feature', inplace=True)
    features.plot(kind='barh', figsize=(50, 50))
    return features
    
tmp = plot_feature_importances(xgb_pipeline_grid2.best_estimator_[2].regressor_, feature_names)
tmp.tail(50)
```
