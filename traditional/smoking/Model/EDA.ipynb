```python
import pandas as pd
import os 
import seaborn as sns
```

```python
dir = '../'
train_data = pd.read_csv(os.path.join(dir, 'train.csv'))
train_data.info()
```

```python
# train_data['age'] = train_data['age'].astype('float')
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "hemoglobin").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "age").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "height(cm)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "weight(kg)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "waist(cm)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "eyesight(left)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "eyesight(right)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "hearing(left)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "hearing(right)").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "systolic").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "relaxation").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "fasting blood sugar").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "Cholesterol").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "triglyceride").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "HDL").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "LDL").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "Urine protein").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "serum creatinine").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "AST").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "ALT").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "Gtp").add_legend()
sns.FacetGrid(train_data, hue="smoking").map(sns.kdeplot, "dental caries").add_legend()
```
