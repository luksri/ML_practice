```python
import pandas as pd
import os 
import seaborn as sns
```

```python
dir = '../'
train_data = pd.read_csv(os.path.join(dir, 'train.csv'))
train_data.info()
```

```python
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "id").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "CustomerId").add_legend()
# sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Surname").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "CreditScore").add_legend()
# sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Geography").add_legend()
# sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Gender").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Age").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Tenure").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "Balance").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "NumOfProducts").add_legend()
# sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "HasCrCard").add_legend()
# sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "IsActiveMember").add_legend()
sns.FacetGrid(train_data, hue="Exited").map(sns.kdeplot, "EstimatedSalary").add_legend()
```
